**.at[row,col] and .iat[row,col]** are optimized for finding a specific element

```python
df.iat[3,5]
```

**filtering with coditions**

```python
df[df["tip"]>1.5]
```

```python
df[(df["size"]>3) & (df["tip"]>1.6)]
```

```python
df[(df["sex"]=="male") | (df["smoker"]=="Yes") & (df["size"]>4)]
```

**Querying with .query()**

```python
df.query("size>1")
```

```python
# use and , or instead of & , |
df.query("size>4 and tip<4")
```

```python
df.query("size>4 and tip<4 or sex=='Male'")
```

```python
# circular query
df.query("3>tip<=7")
```

```python
df.query("3>tip<=7 and size>3")
```

```python
#using variable inside query by making f string 
var="smoker"
var2="day"
df.query(f"{var}=='Yes' and {var2}=='Sat'")
```

```python
# use @ for usina global variable inside " "
bill=18
df.query("total_bill> @bill")
```
