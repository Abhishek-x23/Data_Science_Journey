```python
import pandas as pd
data={
    "name":["Abhishek","Barsha","Ankita"],
    "gender":["male","female","female"],
    "maths":[100,78,56],
    "physics":[99,80,78],
    "chemistry":[98,88,90]
}
df=pd.DataFrame(data)
```

```python
df
```

**melt**

.melt(id_var=[columns staying intact],
        value_vars=[columns that are melting],
        var_name=new name to melted columns,
        value_names=column name for their value)

```python
df2=df.melt(id_vars=["name","gender"],value_vars=["maths","physics","chemistry"],var_name="subject",value_name="marks")
df2
```

**Pivot** opposite of .melt
.pivot_table(index=[],columns="  ",velues="  ")

index: The column whose unique values will become the rows of the new DataFrame.
columns: The column whose unique values will become the columns of the new DataFrame.
values: The column whose values will fill the new DataFrame. These will become the actual data (values in the table).
            
            

```python
df2.pivot(index=["name","gender"],columns="subject",values="marks")
```
