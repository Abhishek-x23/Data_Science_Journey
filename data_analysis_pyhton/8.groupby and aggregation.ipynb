**Group By and Aggregation**

```python
import pandas as pd
```

```python
df = pd.DataFrame({
    "Department": ["HR", "HR", "IT", "IT", "Marketing", "Marketing", "Sales", "Sales"],
    "Team": ["A", "A", "B", "B", "C", "C", "D", "D"],
    "Gender": ["M", "F", "M", "M", "M", "F", "F", "F"],
    "Salary": [85, 90, 78, 85, 92, 88, 80, 80],
    "Age": [23, 25, 30, 22, 28, 26, 21, 27],
    "JoinDate": pd.to_datetime([
        "2020-01-10", "2020-02-15", "2021-03-20", "2021-04-10",
        "2020-05-30", "2020-06-25", "2021-07-15", "2021-08-01"
    ])
})  
```

```python
df
```

```python
df.groupby("Department")["Salary"].mean()
```

```python
df.groupby("Department")["Salary"].sum()
```

```python
df.groupby("Department")["Salary"].min()
```

```python
df.groupby("Department")["Salary"].max()
```

```python
df.groupby("Department")["Salary"].count()
```

```python
df.groupby("Department")[["Salary","Age"]].sum()
```

```python
df.groupby("Department")[["Salary","Age","Gender"]].max()
```

```python
df.groupby("Department")[["Salary","Age","Gender"]].min()
```

```python
df.groupby(["Department","Gender"])[["Salary","Age","Gender"]].max()
```

```python
df.groupby(["Department","Gender"])[["Salary","Age","JoinDate"]].count()
```

```python
df.groupby(["Team","Department"])["Salary"].mean()
```

```python
type(df.groupby(["Team","Department"])["Salary"].mean())
```

__Aggregate__

.agg and .aggregate are same . Both of them return DataFrame

```python
df.groupby(["Department"])["Salary"].agg(["mean","max","count"])
```

```python
df.groupby(["Department"])["Salary"].agg(["mean","max","min","count","sum"])
```

**Filter**

```python
df.groupby(["Department"]).filter(lambda x : x ["Salary"].mean() >85)
```

**Transform**

adds a new column with the returned value of groupby

```python
df["Mean_Salary"]=df.groupby(["Department"])["Salary"].transform("mean")
```

```python
df
```

```python
df["Avg_Gender_Age"]=df.groupby(["Gender"])["Age"].transform("mean")
df
```

```python
df["Max_Dept_Salary"]=df.groupby(["Department"])["Salary"].transform("max")
df
```
