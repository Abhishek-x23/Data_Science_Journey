**creating DataFrames from different datastructures**

```python
import pandas as pd
```

**from list**

```python
#from list
a=[["a",99,100,98],["b",45,90,23],["c",56,87,24]]
df=pd.DataFrame(a)
df
```

```python
b=pd.DataFrame(a,columns=["name","physics","chemistry","maths"])
b
```

**from dictionary**

```python
dict={"abhishek":[99,98,100],"sneha":[34,42,32],"barsha":[80,87,67]}
df2=pd.DataFrame(dict)
df2
```

```python
dic3={"name":["abhishek","sneha","barsha"],"physics":[99,38,70],"chemistry":[98,42,82],"maths":[100,57,87]}
df3=pd.DataFrame(dic3)
df3
```

**from numpy array**

```python
import numpy as np
arr=np.array([[1,2,3],[4,5,6],[4,3,2]])
df4=pd.DataFrame(arr ,  columns=["a","b","c"])
df4
```
