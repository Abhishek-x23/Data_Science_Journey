```python
import pandas as pd
```

```python
df=pd.read_csv("dataset/data2.csv")
df
```

**sorting rows**

```python
# ont the basis of the value of aptitude column
# ascending = False means decreasing order
df.sort_values("aptitude",ascending=False)
```

```python
# sorts on the basis of aptitude and if clash happens then it will chwck DSA
df.sort_values(["aptitude","DSA"],ascending=False)
```

```python
df2=df.sort_values(["DSA","interview"],ascending=False)
df2
```

```python
# this will sort the new rearranged index
df2.sort_index()
```

```python
# this will add new index column while keeping the older one 
# by drop=true we can drop the older index
df2.reset_index()
```

```python
df2.reset_index(inplace=True,drop=True)
df3=df2.copy()
```

```python
df3
```

```python
# makes a new coumns according to the rank of skills column
df3["rank_skill"]=df3["skills"].rank()
df3
```

```python
# normal ranking system
df3["rank_skill"]=df3["skills"].rank(ascending=False)
df3
```

```python
# method = dense gives the same rank to more then one row with same value
df3["rank_skill"]=df3["skills"].rank(ascending=False,method="dense")
df3
```
