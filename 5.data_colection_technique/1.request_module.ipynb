```python
for i in range(1,11):
    
    url=(f"https://quotes.toscrape.com/page/{i}/")
    a= requests.get(url)
    b=a.text
    if a.status_code==200:
        print(f"successfully scraped qute {i}")
    else:
        print(f"failed to scrape quote {i}")
    with open(f"allQuotes/quote{i}.html" ,"w",encoding="utf-8") as f:
        f.write(b)
        
```
