```python
import numpy as np
import time
```

**vectorization** by vectorization the numpyimplements mathematical operation on all the elements simultaneously because it uses SIMD

```python
arr1=[1,2,3,4,5,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4]
start=time.time()
result=[x**2 for x in arr1]
end=time.time()
print(result)
print(end-start)
```

```python
arr1=np.array([1,2,3,4,5,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4,4])
start=time.time()
result=arr1**2
end=time.time()
print(result)
print(end-start)
```

**broadcasting** it means performing a operation with a scalar on the whole array withour using extra space (ex . adding an element in an array ).
It basically streches the smaller data to fit in the bigger one and broadcasts it into bigger one 


```python
a=np.array([1,2,3,4,5])
a
```

```python
a+[10]
```

```python
# 1D array addittion (broadcasting)
arr1=np.array([1,2,3,4,5])
arr2=np.array([1,2,3,4,5])
arr1+arr2
```

```python
# 2D addition (broadcasting)
arr1=np.array([[1,2,3],[4,5,6]])
arr2=np.array([10,20,30])
arr1+arr2
```
