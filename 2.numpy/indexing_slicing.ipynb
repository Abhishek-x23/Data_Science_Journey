```python
import numpy as np
```

```python

a= np.array ([[1,2,3],[6,5,4],[7,8,9]])
```

```python
flat=a.flatten()
```

```python
flat
```

```python
flat[2:]
```

```python
flat[:5]
```

```python
flat[:]
```

```python
# [start:end:gap]
flat[::2]
```

```python
flat[2:7:2]
```

**note:** In numpyslicing gives a view not a copy so If we make changes in the slice it will reflect the original array 
to overcome from this we use .copy() method 

```python
b=flat[2:7]
```

```python
b
```

```python
b[1]
```

```python
b[1]=300
flat
```

```python
b=flat[2:7].copy()
```

```python
b[1]=1000
flat
```

```python
# fancy indexing arr[[index,index...]]
flat[[2,3,5]]
```

```python
# boolean masking arr[condition]
flat[flat>5]
```
