```python
import matplotlib.pyplot as plt
```

```python
study_hours = [1, 2, 3, 4, 5, 6, 7, 8, 9]
exam_scores = [40, 45, 50, 55, 60, 65, 75, 85, 90]
```

```python
plt.scatter(study_hours,exam_scores)
plt.grid()
```

```python
color=["red" if score<50 else "green" for score in exam_scores]

plt.scatter(study_hours,exam_scores,c=color)
plt.grid()
```

```python
color=["red" if score<50 else "green" for score in exam_scores]
size=[score * 2 if score>50 else score for score in exam_scores]
plt.scatter(study_hours,exam_scores,color=color,s=size)
plt.grid()
```

```python
size=[score * 2 if score>50 else score for score in exam_scores]
plt.scatter(study_hours,exam_scores,c=exam_scores,s=size,cmap="viridis")
plt.colorbar(label="color score")
plt.grid()
```

```python
size=[score * 2 if score>50 else score for score in exam_scores]
plt.scatter(study_hours,exam_scores,c=exam_scores,s=size,cmap="viridis")

for i in range(len(study_hours)):
    plt.annotate(f'S{i+1}',(study_hours[i],exam_scores[i]))

plt.colorbar(label="color score")
plt.grid()
```

```python
size=[score * 2 if score>50 else score for score in exam_scores]
plt.scatter(study_hours,exam_scores,c=exam_scores,s=size,cmap="viridis")

for i in range(len(study_hours)):
    plt.annotate(f'S{i+1}',(study_hours[i]+0.3,exam_scores[i]))

plt.colorbar(label="color score")
plt.grid()
```

```python
size=[score * 2 if score>50 else score for score in exam_scores]
plt.scatter(study_hours,exam_scores,c=exam_scores,s=size,cmap="viridis")

for i in range(len(study_hours)):
    plt.annotate(f'S{i+1}',(study_hours[i]+0.1,exam_scores[i]-2))

plt.colorbar(label="color score")
plt.grid()
```

```python
class_a_hours = [2, 4, 6, 8]
class_a_scores = [45, 55, 100, 85]
 
class_b_hours = [1, 3, 5, 7, 9]
class_b_scores = [40, 90, 60, 70, 90]


plt.scatter(class_a_hours,class_a_scores,label="class_a",color="yellow")
plt.scatter(class_b_hours,class_b_scores,label="class_b",color="red")
plt.legend()
plt.grid()
plt.show()
```
