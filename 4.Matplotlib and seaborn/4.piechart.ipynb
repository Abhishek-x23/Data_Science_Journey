# matplotlib Pai Chart

```python
import matplotlib.pyplot as plt
plt.style.use("ggplot")
```

```python
names = ["Sachin", "Sehwag", "Kohli", "Yuvraj"]
runs = [18000, 8000, 12000, 9500]
plt.pie(runs,labels=names)
plt.title("runs")
plt.show()
```

# wedgeprop 
**styling**

```python
colors=["red","blue","green","orange"]
plt.pie(runs,labels=names,
        colors=colors,
       wedgeprops= {'linewidth':2,'edgecolor': "gray",'linestyle':'--'})
plt.title("runs")
plt.show()
```

## explode
### popping a specific element 

```python
explode=[0.1,0.2,0,0]
plt.pie(runs,labels=names,
        colors=colors,
        explode=explode,
       wedgeprops= {'linewidth':2,'edgecolor': "gray",'linestyle':'--'})
plt.title("runs")
plt.show()
```

## shadow

```python
explode=[0.1,0.2,0,0]
plt.pie(runs,labels=names,
        colors=colors,
        explode=explode,
        shadow='true',
       wedgeprops= {'linewidth':2,'edgecolor': "gray",'linestyle':'-'})
plt.title("runs")
plt.show()
```

## Rotation of Pie chart

```python
explode=[0.1,0.2,0,0]
plt.pie(runs,labels=names,
        colors=colors,
        explode=explode,
        startangle=140,
       wedgeprops= {'linewidth':2,'edgecolor': "gray",'linestyle':'-'})
plt.title("runs")
plt.show()
```

## showing percentage

```python
explode=[0.1,0.2,0,0]
plt.pie(runs,labels=names,
        colors=colors,
        explode=explode,
        startangle=140,
        autopct="%1.1f%%",
        shadow="true",
       wedgeprops= {'linewidth':1,'edgecolor': "gray",'linestyle':'--'})
plt.title("runs")
plt.show()
```
