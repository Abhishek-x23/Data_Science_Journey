```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5,6,9,11,13]
```

```python
y1=[i * 2 for i in x]
plt.subplot(1,1,1)
plt.plot(x,y1)
```

```python
y2=[i**2 for i in x]
plt.subplot(1,2,1)
plt.plot(x,y1)

plt.subplot(1,2,2)
plt.plot(x,y2)
```

```python
y3=[i ** 0.5 for i in x]
y4=[i-10 for i in x]
plt.subplot(2,2,1)
plt.plot(x,y1)
plt.title("x*2")

plt.subplot(2,2,2)
plt.plot(x,y2)
plt.title("x**2")

plt.subplot(2,2,3)
plt.plot(x,y3)
plt.title("x**0.5")

plt.subplot(2,2,4)
plt.plot(x,y4)
plt.title("x-10")
```

```python
fig,axs=plt.subplots(1,2,figsize=(6,4))
axs[0].plot(x,y1)
axs[0].set_title("x*2")

axs[1].plot(x,y2)
axs[1].set_title("x**2")

fig.suptitle("plotting",fontsize=14)
fig.tight_layout()
```

```python
fig,axs=plt.subplots(2,2,figsize=(6,4))
axs[0,0].plot(x,y1)
axs[0,0].set_title("x*2")

axs[0,1].plot(x,y2)
axs[0,1].set_title("x**2")

axs[1,0].plot(x,y3)
axs[1,0].set_title("x**0.5")

axs[1,1].plot(x,y4)
axs[1,1].set_title("x-10")

fig.suptitle("plotting",fontsize=14)
fig.tight_layout()
plt.show()
```

```python
fig,axs=plt.subplots(2,2,figsize=(6,4))
n=0
y=[y1,y2,y3,y4]
titles=["y*2","y**2","y**0.5","y-10"]
for i in range(2):
    for j in range(2):
        axs[i][j].plot(x,y[n])
        axs[i][j].set_title(titles[n])
        n+=1
```
