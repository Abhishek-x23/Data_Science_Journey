```python
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
runs =  [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]
plt.plot(years,runs)
plt.xlabel("Years")
plt.ylabel("Runs")
plt.title("Sachin Tendulkar's yearly run")
plt.show()
```

**ploting two set of datas in a single graph**

```python
kohli = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
sehwag = [0, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 0]
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]


plt.plot(years,kohli,label="Kohli runs")
plt.plot(years,sehwag,label="Sehwag runs")
plt.xlabel("years")
plt.ylabel("player's run")
plt.title("kohli and sehwag run comparison")
plt.legend()
plt.show()
```

```python



plt.plot(years,kohli)
plt.plot(years,sehwag)
plt.legend(["Kohli runs","Sehwag runs"])
plt.xlabel("years")
plt.ylabel("player's run")
plt.title("kohli and sehwag run comparison")
plt.show()
```

## styling 
**"color marking linestyle"**

check the official page of matplotlib which are the options available for this 

```python


plt.plot(years,kohli,"ro--",label="Kohli runs")
plt.plot(years,sehwag,"y^:",label="Sehwag runs")
plt.xlabel("years")
plt.ylabel("player's run")
plt.title("kohli and sehwag run comparison")
plt.legend()
plt.show()
```

```python

plt.plot(years,kohli,"gx--",label="Kohli runs")
plt.plot(years,sehwag,"b+:",label="Sehwag runs")
plt.xlabel("years")
plt.ylabel("player's run")
plt.title("kohli and sehwag run comparison")
plt.legend()
plt.show()
```

```python
# they can be written seperately 


plt.plot(years,kohli,color="blue",linestyle="--",marker="*",label="Kohli runs")
plt.plot(years,sehwag,color="green",linestyle="-.",marker="+",label="Sehwag runs")
plt.plot(years,runs,color="red",linestyle=":",marker="^",label="Sachin runs")
plt.xlabel("years")
plt.ylabel("player's run")
plt.title("player's run comparison")
plt.legend()
plt.show()
```

```python
plt.plot(years,kohli,"b*--",label="Kohli runs")
plt.plot(years,sehwag,"g+-.",label="Sehwag runs")
plt.plot(years,runs,"r^:",label="Sachin runs")
plt.xlabel("years")
plt.ylabel("player's run")
plt.title("player's run comparison")
plt.legend()
plt.grid()
plt.show()
```
