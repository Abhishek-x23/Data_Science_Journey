# Histogram

```python
import matplotlib.pyplot as plt
import numpy as np
 
ages = [
    34, 28, 36, 45, 27, 27, 45, 37, 25, 35,
    25, 25, 32, 10, 12, 24, 19, 33, 20, 15,
    44, 27, 30, 15, 24, 31, 18, 33, 23, 27,
    23, 48, 29, 19, 38, 17, 32, 10, 16, 31,
    37, 31, 28, 26, 15, 22, 25, 40, 33, 12,
    33, 26, 23, 36, 40, 39, 21, 26, 33, 39,
    25, 28, 18, 18, 38, 43, 29, 40, 33, 23,
    33, 45, 29, 45, 3, 38, 30, 27, 30, 10,
    27, 33, 44, 24, 21, 24, 39, 33, 24, 35,
    30, 39, 22, 26, 26, 15, 32, 32, 30, 27
]
```

```python
plt.hist(ages,bins=10)
```

```python
plt.hist(ages,bins=10,edgecolor="black")
```

```python
# bin 5 means the data is divided into equal distance bins 
plt.hist(ages,bins=5,edgecolor="black")
```

```python
# with desired bins distance
bins=[10,20,30,40,50,60]
plt.hist(ages,bins=bins,edgecolor="black")
```

```python
bins=[0,10,20,30,40,50]
plt.hist(ages,bins=bins,edgecolor="black")
```

```python
# histogram with random data generated
import random
arr=[]
for x in range (100):
    x=random.randint(0,100)
    arr.append(x)
plt.hist(arr,bins=10,edgecolor="black",alpha=0.7)
plt.xlabel("magnitude of data")
plt.ylabel("frequency of data")
plt.title("frequency of random data generated")
plt.grid()
plt.show()
```

## .axvline(point,...style)


```python

plt.hist(ages,bins=6,color="pink",edgecolor="black")
plt.axvline(np.mean(ages),color="red",linestyle="--",linewidth=2,label="mean")

```

```python
arr=np.array(arr)
plt.hist(arr,bins=10,edgecolor="black",color="pink")

plt.axvline(np.mean(arr),color="red",linestyle="--",linewidth=2,label="mean")

plt.xlabel("magnitude of data")
plt.ylabel("frequency of data")
plt.title("frequency of random data generated")
plt.grid()
plt.legend()
plt.show()
```
