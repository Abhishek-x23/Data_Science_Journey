# Matplotlib Stack Plot

```python
import matplotlib.pyplot as plt
```

```python
days=[1,2,3,4,5,6]
playing=[1,3,5,1,6,1]
studying=[4,5,2,7,8,7]
sleeping=[3,5,2,2,1,0]

plt.figure(figsize=(8,8))
colors=["red","orange","yellow"]
labels=["playing","studying","sleeping"]
plt.stackplot(days,playing,studying,sleeping,labels=labels,colors=colors,alpha=0.8)
plt.legend(loc="upper left")
```
