# barchart in Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
runs =  [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]
```

### simple barchart

```python
plt.bar(years,runs)
plt.xlabel("years")
plt.ylabel("runs")
plt.title("Sachin Tendulkar's Yearly Runs")
plt.show()
```

### barchart with multiple values
**x=np.arange(length)**  - spreads the data in equal gap (0,1,2,3,4,....length)
**plt.xticks(x,years)** - matches the years data with the spreaded data 

```python
sachin = [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]
sehwag = [0, 200, 900, 1400, 1600, 1800, 1500, 1100, 800, 0]
kohli  = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]

x=np.arange(len(years))
width=0.25

plt.bar(x-width,sachin,width=width,label="sachin")
plt.bar(x,sehwag,width=width,label="sehwag")
plt.bar(x+width,kohli,width=width,label="kohli")

# x-width shifts the its bar left side 
# x+width shifts the bar to right side 
# if we dont shift them then all will overlap with each other 
plt.xlabel("years")
plt.ylabel("runs")
plt.title("runs of players")
plt.grid()
plt.legend()
plt.show()
```

```python
sachin = [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]
sehwag = [0, 200, 900, 1400, 1600, 1800, 1500, 1100, 800, 0]
kohli  = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]

x=np.arange(len(years))
width=0.1

plt.bar(x-width,sachin,width=width,label="sachin")
plt.bar(x,sehwag,width=width,label="sehwag")
plt.bar(x+width,kohli,width=width,label="kohli")
plt.xlabel("years")
plt.ylabel("runs")
plt.title("runs of players")
plt.xticks(x,years)
plt.grid()
plt.legend()
plt.show()
```

### without np.arange and xtick

```python
players = ["Sachin", "Sehwag", "Kohli", "Yuvraj"]
runs_5yrs = [500+700+1100+1500+1800, 0+200+900+1400+1600, 0+0+500+800+1100, 300+600+800+1100+900]
```

```python
plt.barh(players,runs_5yrs,color="blue")
plt.xlabel("players")
plt.ylabel("runs_5yrs")
plt.title("Sachin Tendulkar's Yearly Runs")
plt.show()
```
