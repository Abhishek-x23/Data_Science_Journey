**Series** it is like a list in python or similar to dictionary . It is 1D in nature and we can give custom index by default it takes normal index.

```python
import pandas as pd
```

```python
s1=pd.Series([22,34,56,23,99,90])
```

```python
type(s1)
```

```python
s1.dtype
```

```python
s1
```

```python
s2=pd.Series([87,98,99,100,23],index=["c","b","x","a","d"])
s2
```
