# **Merge and Joining**

```python
import pandas as pd
```

```python
employees = pd.DataFrame({
    "EmpID": [1, 2, 3],
    "Name": ["Alice", "Bob", "Charlie"],
    "DeptID": [10, 20, 30]
})

departments = pd.DataFrame({
    "DeptID": [10, 20, 40],
    "DeptName": ["HR", "Engineering", "Marketing"]
})
managers = pd.DataFrame({
    "DeptID": [10, 30, 40,70],
    "Name": ["Ankit", "Abhishek", "Prachi","Sneha"]
})
```

```python
# InnerJoin
pd.merge(employees,departments,on="DeptID")
```

```python
# LeftJoin
pd.merge(employees,departments,on="DeptID", how="left")
```

```python
# RightJoin
pd.merge(employees,departments,on="DeptID",how="right")
```

```python
# OuterJoin
pd.merge(employees,managers,on="DeptID",how="outer")
```

```python
pd.merge(employees,departments,on="DeptID",how="outer")
```

```python
df2=pd.merge(managers,pd.merge(employees,departments,on="DeptID"),on="DeptID",how="right")
df2
```

```python
df2.rename(columns={"Name_x":"Manager Name"},inplace=True)
```

## **Concatination**

```python
df3=pd.DataFrame({
        "Name":["Abhishek","Rahul","Barsha"],
        "Age":[21,20,21]
})
df4=pd.DataFrame({
        "Name":["Anshika","Gaurav","Ankita"],
        "Salary":[8,7,10]
})
df5=pd.DataFrame({
        "Name":["Ashish","Ankita","Sahad"],
        "Age":[22,20,21]
})
```

```python
df3
```

```python
df4
```

```python
pd.concat([df3,df4])
```

```python
pd.concat([df3,df4,df5])
```

```python
df7=pd.concat([df3,df4,df5])
df7.sort_index()
```

axis=1 for width concatination by making new columns

```python
pd.concat([df3,df4],axis=1)
```

```python
df6=pd.concat([df3,df4,df5],axis=1)
df6
```
